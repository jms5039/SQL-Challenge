```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from databaseconfig import user_name, password, local_host
!pip install psycopg2-binary ipykernel
```

```python
# Create a connection string
connection_string = f'postgresql://{user_name}:{password}@{local_host}:5432/EmployeeSQL'
```

```python
# Connect to the PostgreSQL database
engine = create_engine(connection_string)
```

```python
# Load data into DataFrames using SQL queries
titles_query = "SELECT * FROM titles"
employees_query = "SELECT employees.*, titles.title FROM employees JOIN titles ON employees.emp_title_id = titles.title_id"
salaries_query = "SELECT * FROM salaries"
dept_emp_query = "SELECT * FROM dept_emp"
departments_query = "SELECT * FROM departments"

titles = pd.read_sql(titles_query, engine)
employees = pd.read_sql(employees_query, engine)
salaries = pd.read_sql(salaries_query, engine)
dept_emp = pd.read_sql(dept_emp_query, engine)
departments = pd.read_sql(departments_query, engine)
```

```python
# Combine relevant tables for analysis
employee_salary = pd.merge(employees, salaries, on='emp_no', how='inner')
employee_department = pd.merge(employee_salary, dept_emp, on='emp_no', how='inner')
employee_info = pd.merge(employee_department, departments, on='dept_no', how='inner')
```

```python
# Display some basic information about the employees
print("Basic Information about Employees:")
print(employee_info[['emp_no', 'first_name', 'last_name', 'sex', 'hire_date', 'title', 'salary', 'dept_name']])
```

```python
# Visualize the distribution of salaries
plt.figure(figsize=(10, 6))
plt.hist(employee_info['salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Employee Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyze the average salary by department
average_salary_by_dept = employee_info.groupby('dept_name')['salary'].mean().sort_values(ascending=False)
print("\nAverage Salary by Department:")
print(average_salary_by_dept)
```

```python
# Visualize the average salary by department
plt.figure(figsize=(10, 6))
plt.bar(average_salary_by_dept.index, average_salary_by_dept.values, color='green')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```
